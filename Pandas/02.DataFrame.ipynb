## Data 준비
사전 (dict) 자료구조를 통해 생성
예) 슬램덩크 주요 인물 8명에 대한 데이터

```python
data={
    '이름' : ['채치수','정대만', '송태섭','윤대협','강백호','변덕규','황태산','윤대협'],
    '학교' : ['북산고','북산고', '북산고','능남고','북산고','능남고','능남고','능남고'],
    '키' : [198,183,168,187,188, 202,188,190],
    '국어' : [90,40,80,20,75,86,15,17],
    '영어' : [30,100,30,24,75,85,25,60],
    '수학' : [40,70,40,26,85,84,35,15],
    '과학' : [20,40,60,27,95,81,25,10],
    '사회' : [10,30,80,29,65,82,15,20],
    'SW특기' : ['Python','Java','C','C++','PYTHON','C#','','']
}
data
    
```

```python
data['이름']
```

##  DataFrame 객체 생성

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
df
```

# DATA 접근

```python
df['이름']
```

```python
df[['이름','키']] # 대괄호를 2개 써야 한다. 
```

## DataFrame 객체 생성(index 지정)

```python
df = pd.DataFrame(data, index=['1번','2번','3번','4번','5번','6번','7번','8번'])
df
```

## DataFrame 객체 생성(Column 지정)
데이터 중에서 원하는 column 만 선택하거나, 순서변경 가능

```python
df = pd.DataFrame(data, columns=['이름','학교','키'])
df
```
