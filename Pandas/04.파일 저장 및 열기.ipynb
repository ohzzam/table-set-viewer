# 4. 파일 저장 및 열기
DataFrame 객체를 excel,csv,txt 등의 형태의 파일로 읽고 쓰기

```python
import pandas as pd
data={
    '이름' : ['채치수','정대만', '송태섭','윤대협','강백호','변덕규','황태산','윤대협'],
    '학교' : ['북산고','북산고', '북산고','능남고','북산고','능남고','능남고','능남고'],
    '키' : [198,183,168,187,188, 202,188,190],
    '국어' : [90,40,80,20,75,86,15,17],
    '영어' : [30,100,30,24,75,85,25,60],
    '수학' : [40,70,40,26,85,84,35,15],
    '과학' : [20,40,60,27,95,81,25,10],
    '사회' : [10,30,80,29,65,82,15,20],
    'SW특기' : ['Python','Java','C','C++','PYTHON','C#','','']
}
data
df = pd.DataFrame(data, index=['1번','2번','3번','4번','5번','6번','7번','8번'])
df.index.name='지원번호'
df
```

## 저장하기 CSV 파일로 저장

```python
df.to_csv('score.csv',encoding='utf-8-sig')  # 인코딩.. 중요.. 
```

```python
df.to_csv('score.csv',encoding='utf-8-sig', index=False) 
```

## Txt 파일로 저장

```python
df.to_csv('score.txt', sep='\t')  # tab 로 구분된 텍스트 파일. 
```

## 엑셀 파일로 저장

```python
df.to_excel('score.xlsx')
```

## 열기
### CSV 파일 열기

```python
df = pd.read_csv('score.csv')
df
```

```python
df = pd.read_csv('score.csv', skiprows=1)
df
```

```python
df = pd.read_csv('score.csv', skiprows=[1,3,5])
df
```

```python
df = pd.read_csv('score.csv', nrows=4) # 지정된 개수 만큼 가져온다.
df
```

```python
df = pd.read_csv('score.csv', skiprows=2, nrows=4) # 처음 2개 row 무시, 이후 4개 가져온다.
df
```

## 텍스트 파일 열기

```python
df = pd.read_csv('score.txt', sep='\t')
df
```

```python
df = pd.read_csv('score.txt', sep='\t', index_col='지원번호')  #, index_col='지원번호'
df.set_index('키', inplace=True)
df
```
